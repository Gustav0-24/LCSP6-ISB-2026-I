# **LABORATORIO 3: Codigos de las Señales EMG con BITalino**

## **EMG - Biceps Braquial**

### 1. Reposo

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt, welch, iirnotch

# ==============================
# Función filtro pasa-banda EMG
# ==============================
def filtro_pasabanda(senal, f_muestreo, frec_baja=20.0, frec_alta=450.0, orden=4):
    nyquist = 0.5 * f_muestreo
    bajo = frec_baja / nyquist
    alto = frec_alta / nyquist
    b, a = butter(orden, [bajo, alto], btype="band")
    return filtfilt(b, a, senal)

# ==============================
# Función filtro notch (60 Hz)
# ==============================
def filtro_notch(senal, f_muestreo, frec_notch=60.0, Q=30.0):
    nyquist = 0.5 * f_muestreo
    w0 = frec_notch / nyquist
    b, a = iirnotch(w0, Q)
    return filtfilt(b, a, senal)

# ==============================
# Cargar archivo (YA SUBIDO)
# ==============================
ruta = "biceps_reposo.txt"   # archivo en el mismo directorio del notebook
datos = np.loadtxt(ruta, comments="#")

# Señal EMG (última columna)
emg_reposo = datos[:, -1]

# ==============================
# Variables de la señal
# ==============================
fs = 1000  # Hz
tiempo_reposo = np.linspace(0, len(emg_reposo) / fs, len(emg_reposo))

# ==============================
# Señal EMG original
# ==============================
plt.figure(figsize=(10, 4))
plt.plot(tiempo_reposo, emg_reposo, linewidth=0.4)
plt.title("Señal EMG original - Bíceps (reposo)")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud (uV)")
plt.grid()
plt.show()

# ==============================
# Filtrado de la señal
# ==============================
emg_reposo_filtrada = filtro_pasabanda(emg_reposo, fs)
emg_reposo_filtrada = filtro_notch(emg_reposo_filtrada, fs, frec_notch=60.0, Q=30.0)

# ==============================
# Señal EMG filtrada
# ==============================
plt.figure(figsize=(10, 4))
plt.plot(tiempo_reposo, emg_reposo_filtrada, linewidth=0.4)
plt.title("Señal EMG filtrada - Bíceps (reposo)")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud (uV)")
plt.grid()
plt.show()

# ==============================
# Welch (PSD)
# ==============================
frecs_reposo, psd_reposo = welch(emg_reposo_filtrada, fs, nperseg=1024)

plt.figure(figsize=(10, 4))
plt.semilogy(frecs_reposo, psd_reposo)
plt.title("Densidad espectral de potencia (Welch) - Bíceps (reposo)")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("PSD (uV²/Hz)")
plt.xlim(0, 500)
plt.grid()
plt.show()

# ==============================
# FFT
# ==============================
fft_vals_reposo = np.fft.rfft(emg_reposo_filtrada)
fft_freqs_reposo = np.fft.rfftfreq(len(emg_reposo_filtrada), 1/fs)

plt.figure(figsize=(10, 4))
plt.plot(fft_freqs_reposo, np.abs(fft_vals_reposo), linewidth=0.6)
plt.title("Espectro de frecuencias (FFT) - Bíceps (reposo)")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Amplitud")
plt.xlim(0, 500)
plt.grid()
plt.show()
```

### 2. Movimiento fuerte con oposición

```python
## 2. Movimiento fuerte con oposición

import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt, welch, iirnotch

# ==============================
# Función filtro pasa-banda EMG
# ==============================
def filtro_pasabanda(senal, f_muestreo, frec_baja=20.0, frec_alta=450.0, orden=4):
    nyquist = 0.5 * f_muestreo
    bajo = frec_baja / nyquist
    alto = frec_alta / nyquist
    b, a = butter(orden, [bajo, alto], btype="band")
    return filtfilt(b, a, senal)

# ==============================
# Función filtro notch (60 Hz)
# ==============================
def filtro_notch(senal, f_muestreo, frec_notch=60.0, Q=30.0):
    nyquist = 0.5 * f_muestreo
    w0 = frec_notch / nyquist
    b, a = iirnotch(w0, Q)
    return filtfilt(b, a, senal)

# ==============================
# Cargar archivo (YA SUBIDO)
# ==============================
ruta = "biceps_oposicion.txt"
datos = np.loadtxt(ruta, comments="#")

# Señal EMG (última columna)
emg_oposicion = datos[:, -1]

# ==============================
# Variables de la señal
# ==============================
fs = 1000  # Hz
tiempo_oposicion = np.linspace(0, len(emg_oposicion) / fs, len(emg_oposicion))

# ==============================
# Señal EMG original
# ==============================
plt.figure(figsize=(10, 4))
plt.plot(tiempo_oposicion, emg_oposicion, linewidth=0.4)
plt.title("Señal EMG original - Bíceps (movimiento fuerte con oposición)")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud (uV)")
plt.grid()
plt.show()

# ==============================
# Filtrado de la señal
# ==============================
emg_oposicion_filtrada = filtro_pasabanda(emg_oposicion, fs)
emg_oposicion_filtrada = filtro_notch(
    emg_oposicion_filtrada, fs, frec_notch=60.0, Q=30.0
)

# ==============================
# Señal EMG filtrada
# ==============================
plt.figure(figsize=(10, 4))
plt.plot(tiempo_oposicion, emg_oposicion_filtrada, linewidth=0.4)
plt.title("Señal EMG filtrada - Bíceps (movimiento fuerte con oposición)")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud (uV)")
plt.grid()
plt.show()

# ==============================
# Welch (PSD)
# ==============================
frecs_oposicion, psd_oposicion = welch(
    emg_oposicion_filtrada, fs, nperseg=1024
)

plt.figure(figsize=(10, 4))
plt.semilogy(frecs_oposicion, psd_oposicion)
plt.title(
    "Densidad espectral de potencia (Welch) - Bíceps (movimiento fuerte con oposición)"
)
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("PSD (uV²/Hz)")
plt.xlim(0, 500)
plt.grid()
plt.show()

# ==============================
# FFT
# ==============================
fft_vals_oposicion = np.fft.rfft(emg_oposicion_filtrada)
fft_freqs_oposicion = np.fft.rfftfreq(
    len(emg_oposicion_filtrada), 1 / fs
)

plt.figure(figsize=(10, 4))
plt.plot(fft_freqs_oposicion, np.abs(fft_vals_oposicion), linewidth=0.6)
plt.title(
    "Espectro de frecuencias (FFT) - Bíceps (movimiento fuerte con oposición)"
)
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Amplitud")
plt.xlim(0, 500)
plt.grid()
plt.show()
```
